#### **2. 함수로 만들기**

#### 1) 함수를 시간대별로 여러번 호출하기



```python
def hourly(hour):
    params ={'key' : key, 'type' :'xml', 'sumTmUnitTypeCode':'3', 
             'stdHour' : hour, 'numOfRows' : '1000', 'pageNo' : '10'}    
    response = requests.get(url, params=params)
    data = str(response.content,"utf-8")
    tree = ET.fromstring(data)
    df = []
    
    for a in range(3, 25):
        temp = []
        # 4번(입구출구), 6번(기준시), 10번(차종), 11번(톨게이트이름), 13번(교통량) 열사용: 위와 비교하여 6번이 추가되었어요
        for b in [4, 6, 10, 11, 13]:     
            temp.append(tree[a][b].text)
        df.append(temp)
    df = pd.DataFrame(df) 

    return df
```

```python
df5 = hourly('05')
df5.head(3)
```

```python
df6 = hourly('06')
df6.head(3)
```

```python
df7 = hourly('07')
df7.head(3)
```

```python
# df 합치기
df = pd.concat([df5, df6, df7])
df
```

#### 2) 여러 시간대를 돌리는 함수 만들기

```python
def many_hours(hour_list):
    dfs = pd.DataFrame([])

    for i in hour_list:   
        df = hourly(i)
        dfs = pd.concat([dfs, df])
    
    return dfs
```

```python
# 뽑을 시간대의 목록을 여기에 입력
hour_list = ['05', '06', '07', '08', '09', '10', '11', '15', '16'] 

df = many_hours(hour_list)
```
