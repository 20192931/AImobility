```python
rs.port_open("COM12") #포트번호는 개인마다 다름
```

```python
#6. 모빌리티 동작 함수 정의하기
# 전진
def forward(speed=5):
    rs.motor_write(0, 1, speed)
    rs.motor_write(1, 0, speed)
    rs.motor_write(2, 1, speed)
    rs.motor_write(3, 0, speed)
    time.sleep(0.1)                                

# 후진    
def reverse():
    rs.motor_write(0, 0, 15)
    rs.motor_write(1, 1, 15)
    rs.motor_write(2, 0, 15)
    rs.motor_write(3, 1, 15)
    time.sleep(0.1)

# 정지
def stop():
    rs.motor_write(0, 1, 0)
    rs.motor_write(1, 0, 0)
    rs.motor_write(2, 1, 0)
    rs.motor_write(3, 0, 0)
    time.sleep(0.1)

# 우측
def right(speed):
    rs.motor_write(0, 1, speed)
    rs.motor_write(1, 1, speed)
    rs.motor_write(2, 0, speed)
    rs.motor_write(3, 0, speed)
    time.sleep(0.1)
    
# 좌측 
def left(speed = 1):
    rs.motor_write(0, 0, speed)
    rs.motor_write(1, 0, speed)
    rs.motor_write(2, 1, speed)
    rs.motor_write(3, 1, speed)
    time.sleep(0.1)
    
#우회전
def R_turn():
    rs.motor_write(0, 1, 10)
    rs.motor_write(1, 1, 10)
    rs.motor_write(2, 1, 10)
    rs.motor_write(3, 1, 10)
    time.sleep(0.1)    
    
def R1_turn():
    rs.motor_write(0, 1, 10)
    rs.motor_write(1, 1, 4)
    rs.motor_write(2, 1, 10)
    rs.motor_write(3, 1, 4)
    time.sleep(0.1)  

#좌회전
def L_turn():
    rs.motor_write(0, 0, 4)
    rs.motor_write(1, 0, 10)
    rs.motor_write(2, 0, 4)
    rs.motor_write(3, 0, 10)
    time.sleep(0.1) 

#한쪽 바퀴  -- 
'''def RR():
    rs.motor_write(0, 1, 15)
    rs.motor_write(1, 0, 0)
    rs.motor_write(2, 1, 15)
    rs.motor_write(3, 0, 0)
    time.sleep(0.1)
def LL():
    rs.motor_write(0, 1, 0)
    rs.motor_write(1, 0, 15)
    rs.motor_write(2, 1, 0)
    rs.motor_write(3, 0, 15)
    time.sleep(0.1)'''
```
